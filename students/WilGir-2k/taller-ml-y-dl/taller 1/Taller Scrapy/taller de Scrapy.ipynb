```python
from bs4 import BeautifulSoup
import requests
```

```python
start_url = 'https://rpp.pe/buscar?q=ingenieria'
domain = 'https://rpp.pe/'
```

```python
def get_soup(url):
    content = requests.get(url).content
    return BeautifulSoup(content,'lxml')
```

```python
''' return a list of links to other wiki articles '''
def extract_articles(url=start_url):
    soup = get_soup(url)
    p_tags = soup.findAll('article')
    a_tags = []
    for p_tag in p_tags:
        a_tags.extend(p_tag.select('h2 a'))

    a_tags = [ a_tag for a_tag in a_tags if 'href' in a_tag.attrs ]

    links = [ a_tag.get('href') for a_tag in a_tags ]
    return links
```

```python
def extract_noticias(url):
    soup = get_soup(url)
    titulo = soup.find('h1').text
    articulos = soup.findAll('p')
    return titulo, articulos
```

```python
if __name__ == '__main__':
    ext_links = extract_articles(start_url)
    noticias=[]
    with open('noticias_ingenieros_rppxx.txt','w',encoding='utf-8-sig') as f:
        for noti in ext_links:
            titulo,contenido=extract_noticias(noti)
            f.write(str(contenido)+'\n')
```
