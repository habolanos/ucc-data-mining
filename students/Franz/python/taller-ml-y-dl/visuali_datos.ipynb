```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('chicago-Crimes-Map.csv')

# Histograma de una variable
plt.hist(data['BLOCK'], bins=20)
plt.xlabel('x')  # Corrección: xlabel en lugar de xlabe
plt.ylabel('y')
plt.title('Histograma de variable')
plt.show()

# Diagrama de dispersión 
plt.scatter(data['BLOCK'], data['BLOCK'])
plt.xlabel('x')
plt.ylabel('y')  # Corrección: ylabel en lugar de xlabel
plt.title('Diagrama de dispersión')
plt.show()
```
